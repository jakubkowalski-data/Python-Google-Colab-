```python
import pandas as pd


url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'

column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income']

df = pd.read_csv(url, header=None, names=column_names)

print("Adult dataset loaded successfully.")
```

```python
print("First 5 rows of the DataFrame:")
print(df.head())

print("\nDataFrame Info:")
df.info()
```

```python
import numpy as np

initial_rows = df.shape[0]
print(f"Initial number of rows: {initial_rows}")

df.replace('?', np.nan, inplace=True)
df_cleaned = df.dropna()

removed_rows = initial_rows - df_cleaned.shape[0]
print(f"Number of rows removed: {removed_rows}")

df = df_cleaned.copy()
print("Missing values replaced with NaN and rows with NaN values removed successfully.")
```

```python
print("Unique values in 'income' column:")
print(df['income'].unique())

print("\nData type of 'income' column:")
print(df['income'].dtype)
```

```python
import pandas as pd

sex_counts = df['sex'].value_counts()
print("Podział według płci:")
print(sex_counts)

male_avg_age = df[df['sex'].str.strip() == 'Male']['age'].mean()
print(f"\nŚredni wiek mężczyzn: {male_avg_age:.2f}")

poland_count = df[df['native-country'].str.strip() == 'Poland'].shape[0]
total_rows = df.shape[0]
poland_percentage = (poland_count / total_rows) * 100
print(f"Procentowy udział osób z Polski: {poland_percentage:.2f}%")

higher_education = ['Bachelors', 'Masters', 'Doctorate', 'Prof-school']

no_higher_edu_high_income = df[
    (~df['education'].str.strip().isin(higher_education)) &
    (df['income'].str.strip() == '>50K')
]

print(f"\nLiczba osób bez wyższego wykształcenia z dochodem >50K: {len(no_higher_edu_high_income)}")
print("Fragment danych (5 wierszy):")
print(no_higher_edu_high_income.head())
```

```python
import pandas as pd

df['education_clean'] = df['education'].str.strip()
age_stats_by_edu = df.groupby('education_clean')['age'].describe()
print("Statystyki wieku według poziomu wykształcenia:")
print(age_stats_by_edu)

married_statuses = ['Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse']

males_df = df[df['sex'].str.strip() == 'Male'].copy()

males_df['marital_group'] = males_df['marital-status'].str.strip().apply(
    lambda x: 'Married' if x in married_statuses else 'Unmarried'
)

earnings_structure = males_df.groupby(['marital_group', 'income']).size().unstack(fill_value=0)

earnings_pct = earnings_structure.div(earnings_structure.sum(axis=1), axis=0) * 100

print("\nStruktura zarobków mężczyzn ze względu na stan cywilny (liczbowo):")
print(earnings_structure)

print("\nStruktura zarobków mężczyzn ze względu na stan cywilny (procentowo):")
print(earnings_pct.round(2).astype(str) + '%')

df.drop(columns=['education_clean'], inplace=True)
```

```python
max_hours = df['hours-per-week'].max()
max_hours_workers = df[df['hours-per-week'] == max_hours]
num_max_workers = len(max_hours_workers)

print(f"Maximum number of working hours per week: {max_hours}")
print(f"Number of individuals working {max_hours} hours per week: {num_max_workers}")
```
