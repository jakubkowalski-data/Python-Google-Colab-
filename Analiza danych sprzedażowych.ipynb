```python
from google.cloud import bigquery

client = bigquery.Client(project='data-analytics-mate')
```

```python
query = """
SELECT
    DATE(s.date) AS order_date,
    s.ga_session_id AS session_id,

    sp.continent,
    sp.country,
    sp.device,
    sp.browser,
    sp.mobile_model_name,
    sp.operating_system,
    sp.language AS browser_language,
    sp.medium AS traffic_source,
    sp.channel AS traffic_channel,

    a.id AS user_id,
    a.is_verified AS is_email_confirmed,
    a.is_unsubscribed AS is_newsletter_unsubscribed,

    p.category AS product_category,
    p.name AS product_name,
    p.price,
    p.short_description

FROM `data-analytics-mate.DA.session` s

LEFT JOIN `data-analytics-mate.DA.session_params` sp
    ON s.ga_session_id = sp.ga_session_id

LEFT JOIN `data-analytics-mate.DA.account_session` acs
    ON s.ga_session_id = acs.ga_session_id

LEFT JOIN `data-analytics-mate.DA.account` a
    ON acs.account_id = a.id

LEFT JOIN `data-analytics-mate.DA.order` o
    ON s.ga_session_id = o.ga_session_id

LEFT JOIN `data-analytics-mate.DA.product` p
    ON o.item_id = p.item_id
"""
```

```python
df = client.query(query).to_dataframe()
display(df.head())
```

```python
import pandas as pd
import numpy as np

df['order_date'] = pd.to_datetime(df['order_date'])

num_cols = len(df.columns)
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object', 'string']).columns.tolist()
date_cols = df.select_dtypes(include=['datetime64', 'datetime']).columns.tolist()

unique_sessions = df['session_id'].nunique()
min_date = df['order_date'].min()
max_date = df['order_date'].max()

missing_values = df.isnull().sum()

print(f"Całkowita liczba kolumn: {num_cols}")
print(f"Liczba kolumn numerycznych: {len(numeric_cols)} ({', '.join(numeric_cols) if numeric_cols else 'brak'})")
print(f"Liczba kolumn kategorycznych: {len(categorical_cols)} ({', '.join(categorical_cols)})")
print(f"Liczba kolumn daty: {len(date_cols)} ({', '.join(date_cols)})")
print(f"Liczba unikalnych sesji: {unique_sessions}")
print(f"Objęty okres: od {min_date} do {max_date}")
print(f"\nBrakujące wartości:\n{missing_values[missing_values > 0]}")

print("\nKomentarz dotyczący braków danych:")
print("Najwięcej brakujących wartości występuje w kolumnach związanych z kontem użytkownika oraz danymi produktu.")
print("Wynika to z faktu, że użyliśmy LEFT JOIN, a sesji zarejestrowanych użytkowników lub sesji zakończonych zamówieniem konkretnego produktu jest relatywnie mało w stosunku do wszystkich sesji (gości).")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
sns.set_theme(style="whitegrid")
```

```python
daily_sales = df.groupby('order_date')['price'].sum().reset_index()

plt.figure(figsize=(15, 6))
sns.lineplot(data=daily_sales, x='order_date', y='price')
plt.title('Ogólna dynamika sprzedaży (Listopad 2020 - Styczeń 2021)')
plt.ylabel('Suma sprzedaży')
plt.show()

plt.figure(figsize=(15, 6))
continent_sales = df.groupby(['order_date', 'continent'])['price'].sum().reset_index()
sns.lineplot(data=continent_sales, x='order_date', y='price', hue='continent')
plt.title('Sprzedaż według kontynentów')
plt.show()
```

```python
# Dynamika wg kanałów i urządzeń
fig, axes = plt.subplots(2, 1, figsize=(15, 12))

channel_sales = df.groupby(['order_date', 'traffic_channel'])['price'].sum().reset_index()
sns.lineplot(data=channel_sales, x='order_date', y='price', hue='traffic_channel', ax=axes[0])
axes[0].set_title('Sprzedaż według kanałów ruchu')

device_sales = df.groupby(['order_date', 'device'])['price'].sum().reset_index()
sns.lineplot(data=device_sales, x='order_date', y='price', hue='device', ax=axes[1])
axes[1].set_title('Sprzedaż według urządzeń')

plt.tight_layout()
plt.show()
```

```python
# Sesje wg kanałów i urządzeń
pivot_sessions = df.dropna(subset=['traffic_channel', 'device']).pivot_table(
    index='traffic_channel', columns='device', values='session_id', aggfunc='nunique')

# Top 10 kategorii i Top 5 krajów
top_categories = df['product_category'].value_counts().nlargest(10).index
top_countries = df['country'].value_counts().nlargest(5).index

pivot_geo_prod = df[df['product_category'].isin(top_categories) & df['country'].isin(top_countries)].pivot_table(
    index='product_category', columns='country', values='price', aggfunc='sum')

display("Tabela: Sesje wg kanału i urządzenia", pivot_sessions)
display("Tabela: Sprzedaż Top Kategorii w Top Krajach", pivot_geo_prod)
```

```python
daily_metrics = df.groupby('order_date').agg({'session_id': 'nunique', 'price': 'sum'})
corr, p_val = stats.pearsonr(daily_metrics['session_id'], daily_metrics['price'])

plt.figure(figsize=(8, 6))
sns.regplot(data=daily_metrics, x='session_id', y='price')
plt.title(f'Korelacja Sesje vs Sprzedaż: {corr:.2f} (p={p_val:.4f})')
plt.show()

# Korelacja między kontynentami (Americas, Europe, Asia)
cont_pivot = df.pivot_table(index='order_date', columns='continent', values='price', aggfunc='sum')[['Americas', 'Europe', 'Asia']].fillna(0)
display("Macierz korelacji między kontynentami:", cont_pivot.corr())
```

```python
# Zarejestrowani vs Goście
df['is_registered'] = df['user_id'].notnull()
sales_reg = df[df['is_registered']].groupby('order_date')['price'].sum()
sales_guest = df[~df['is_registered']].groupby('order_date')['price'].sum()

t_stat, t_p = stats.ttest_ind(sales_reg, sales_guest)
print(f"Test T dla zarejestrowanych vs gości: p-value = {t_p:.4f}")

# Proporcja Organic Search: Europe vs Americas
cont_table = pd.crosstab(df['continent'].isin(['Europe', 'Americas']), df['traffic_channel'] == 'Organic Search')
chi2, p_chi2, _, _ = stats.chi2_contingency(cont_table)
print(f"Test Chi2 dla ruchu organicznego (Europe vs Americas): p-value = {p_chi2:.4f}")
```

```python
from statsmodels.stats.proportion import proportions_ztest

# Porównanie proporcji sesji z zakupem: Desktop vs Mobile
desktop_sessions = df[df['device'] == 'desktop']
mobile_sessions = df[df['device'] == 'mobile']

successes = np.array([
    desktop_sessions['price'].notnull().sum(),
    mobile_sessions['price'].notnull().sum()
])
nobs = np.array([
    len(desktop_sessions),
    len(mobile_sessions)
])

z_stat, p_ztest = proportions_ztest(successes, nobs)

print("\n--- Test proporcji (Konwersja: Desktop vs Mobile) ---")
conv_desktop = (successes[0] / nobs[0]) * 100
conv_mobile = (successes[1] / nobs[1]) * 100

print(f"Konwersja Desktop: {conv_desktop:.2f}%")
print(f"Konwersja Mobile: {conv_mobile:.2f}%")
print(f"p-value: {p_ztest:.4f}")

if p_ztest < 0.05:
    print("Wynik: Różnica w proporcji konwersji między urządzeniami jest statystycznie istotna.")
else:
    print("Wynik: Brak statystycznie istotnych różnic w proporcji konwersji.")
```

```python
from scipy.stats import f_oneway

sales_data = df.dropna(subset=['price'])

group_americas = sales_data[sales_data['continent'] == 'Americas']['price']
group_europe = sales_data[sales_data['continent'] == 'Europe']['price']
group_asia = sales_data[sales_data['continent'] == 'Asia']['price']

f_stat, p_anova = f_oneway(group_americas, group_europe, group_asia)

print("--- Analiza ANOVA (Porównanie średniej wartości zakupu między kontynentami) ---")
print(f"F-statistic: {f_stat:.4f}")
print(f"p-value: {p_anova:.4f}")

if p_anova < 0.05:
    print("Wynik: Istnieją statystycznie istotne różnice w średniej wartości zakupu między wybranymi kontynentami.")
else:
    print("Wynik: Brak statystycznie istotnych różnic w średniej wartości zakupu między kontynentami.")
```
